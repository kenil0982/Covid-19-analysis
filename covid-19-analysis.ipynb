```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

dataset=pd.read_csv("/kaggle/input/covid-data/covid19_Confirmed_dataset.csv")
dataset.head()
df= dataset.drop(["Lat","Long"],axis=1,inplace=True)
corona_dataset_aggregated=dataset.groupby ("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection = []

for c in countries:
    # Get the row data (time series) for the country
    data = corona_dataset_aggregated.loc[c]
    
    # Ensure data is numeric (in case of any strings)
    data_numeric = pd.to_numeric(data, errors='coerce')
    
    # Calculate daily differences and find the maximum
    max_val = data_numeric.diff().max()
    
    max_infection.append(max_val)

# Add new column to DataFrame (requires resetting index or storing separately)
corona_dataset_aggregated["Max Infection"] = max_infection
corona_dataset_aggregated
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Happiness Dataset
happiness = pd.read_csv("/kaggle/input/covid-data/worldwide_happiness_report.csv")

# 2. Prepare corona_dataset_aggregated (Assuming you've already added 'Max Infection' to it)
# Rename the index to 'Country' to align with happiness dataset
corona_df = corona_dataset_aggregated.copy()
corona_df["Country"] = corona_df.index

# 3. Rename column in happiness to match for merge
happiness.rename(columns={"Country or region": "Country"}, inplace=True)

# 4. Merge DataFrames on 'Country'
merged = pd.merge(happiness, corona_df[["Country", "Max Infection"]], on="Country", how="inner")

# 5. Drop missing values if any
merged.dropna(subset=["GDP per capita", "Max Infection"], inplace=True)

# 6. Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged, x="GDP per capita", y="Max Infection")
plt.title("GDP per Capita vs Max Daily COVID Infections")
plt.xlabel("GDP per Capita")
plt.ylabel("Max Daily Infections")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'merged' DataFrame has already been created from previous steps
x = merged["GDP per capita"]
y = merged["Max Infection"]

# Remove non-positive y values before applying log
mask = y > 0
x = x[mask]
y = y[mask]

# Plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x=x, y=np.log(y))
plt.xlabel("GDP per Capita")
plt.ylabel("Log of Max Daily Infections")
plt.title("GDP vs Log(Max Infection) with Regression Line")
plt.grid(True)
plt.show()
```
