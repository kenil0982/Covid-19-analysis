```python
countries = list(corona_dataset_aggregated.index)
max_infection = []

for c in countries:
    # Get the row data (time series) for the country
    data = corona_dataset_aggregated.loc[c]
    
    # Ensure data is numeric (in case of any strings)
    data_numeric = pd.to_numeric(data, errors='coerce')
    
    # Calculate daily differences and find the maximum
    max_val = data_numeric.diff().max()
    
    max_infection.append(max_val)

# Add new column to DataFrame (requires resetting index or storing separately)
corona_dataset_aggregated["Max Infection"] = max_infection
```

```python
corona_dataset_aggregated
```

```python
corona_date=pd.DataFrame(corona_dataset_aggregated["Max Infection"])
```
